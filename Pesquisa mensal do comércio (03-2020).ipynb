```python
import numpy as np
import chart_studio
import pandas as pd
import plotly.graph_objs as go
import chart_studio.plotly as py


from plotly.subplots import make_subplots

chart_studio.tools.set_config_file(world_readable = True, sharing = 'public')
```

<div class="alert alert-block alert-info">
<b>Carregar dados sobre comércío</b>
</div>

```python
comercio = pd.read_excel('Pesquisa mensal do comércio (03-2020).xlsx', sheet_name = 'Comércio')
comercio.set_index('Brasil e Unidades da Federação', inplace = True)
comercio = comercio.sort_values(by = ['Comércio varejista ampliado'], ascending = True)
comercio
```

<div class="alert alert-block alert-info">
<b>Carregar dados sobre atividade comércial</b>
</div>

```python
atividade = pd.read_excel('Pesquisa mensal do comércio (03-2020).xlsx', sheet_name = 'Atividades')
atividade.set_index('Atividades', inplace = True)
atividade.sort_index(inplace = True)
atividade
```

<div class="alert alert-block alert-info">
<b>Plot sobre comércío varejista e comércío varejista ampliado</b>
</div>

```python
fig = go.Figure()

fig.add_trace(go.Bar(x = comercio.index, y = comercio['Comércio varejista'], 
    name = 'Comércio varejista',
    hovertemplate = '%{label} <br>Acumulado: %{y}',
    text = comercio['Comércio varejista'],
    textposition = 'outside',
    marker_color = 'rgb(255, 150, 50)'))

fig.add_trace(go.Bar(x = comercio.index, y = comercio['Comércio varejista ampliado'], 
    name = 'Comércio varejista ampliado',
    hovertemplate = '%{label} <br>Acumulado: %{y}',
    text = comercio['Comércio varejista ampliado'],
    textposition = 'outside',
    marker_color = 'rgb(255, 255, 50)'))

fig.update_layout(title_text = 'Comércio varejista (Jan-Mar) <br> Acumulado do ano',
    yaxis_title = 'Variação (%) do volume de vendas do comércio',
    xaxis_title = 'Brasil e estados selecionados',
    title_x = 0.5,
    title_y = 0.95,
    autosize = False,
    height = 720,
    width = 1280,
    barmode = 'group',
    bargap = 0.2, 
    bargroupgap = 0.2,            
    legend = dict(
    x = 0, y = 1.2),
    paper_bgcolor = 'rgba(1,1,1,1)',
    plot_bgcolor = 'rgba(1,1,1,1)',
    hoverlabel = dict(
    bgcolor = 'white',
    font_color = 'black',
    font_size = 15, 
    font_family = 'Helvetica'),
    font = dict(
    family = 'Helvetica, Helvetica',
    size = 15,
    color = 'white'))

fig.update_xaxes(tickangle = 315,
                 ticks = 'outside', 
                 tickwidth = 2, 
                 tickcolor = 'black', 
                 ticklen = 10,                  
                 showgrid = False)

fig.update_yaxes(ticks = 'outside', 
                 tickwidth = 2, 
                 tickcolor = 'black', 
                 ticklen = 10,              
                 zeroline = False)

py.iplot(fig, filename = 'Comércio varejista (03-2020) usando acumulado do ano', auto_open = False)
```
